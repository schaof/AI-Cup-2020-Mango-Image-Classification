# test

```
os.chdir("/content/drive/My Drive/P2_test/Test") #test圖片原儲存位置
for ii in range(len(test)): 
    image=cv.imread(test.image_id[ii])
    x=test.pos_x[ii]
    if x<0:
      x=0
    y=test.pos_y[ii]
    if y<0:
      y=0
    w=test.width[ii]
    h=test.height[ii]
    crop=image[y:y+h,x:x+w]
    cv.imwrite('/content/drive/My Drive/P2_test/Test_seg/'+test.image_id[ii], crop, [cv.IMWRITE_JPEG_QUALITY, 100])
```
